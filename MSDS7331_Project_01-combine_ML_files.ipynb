# Combine Data
This section will combine the ML ready data sets and add a year column indicating which year's file the observations originated from. 

```python
#### Reference: Snippets of this code block was taken directly from Jake Drew git repository: https://github.com/jakemdrew ####

#import required Libraries
import pandas as pd
import numpy as np
import os
import string

#**********************************************************************************
# Set the following variables before running this code!!!
#**********************************************************************************

#Location where copies of the raw data files will be downloaded and saved as csv files.
dataDir = 'C:/Users/flip/Desktop/SMU/MSDS7331/data/'
```

```python
#### Reference: Snippets of this code block was taken directly from Eric Larson git repository: https://github.com/eclarson ####
#### Reference: https://stackoverflow.com/questions/12555323/adding-new-column-to-existing-dataframe-in-python-pandas

df_2014hsML = pd.read_csv(dataDir + 'PublicHighSchools2014_ML.csv') # read in the csv file
df_2014hsML['year'] = 2014

df_2015hsML = pd.read_csv(dataDir + 'PublicHighSchools2015_ML.csv') # read in the csv file 
df_2015hsML['year'] = 2015

df_2016hsML = pd.read_csv(dataDir + 'PublicHighSchools2016_ML.csv') # read in the csv file 
df_2016hsML['year'] = 2016

df_2017hsML = pd.read_csv(dataDir + 'PublicHighSchools2017_ML.csv') # read in the csv file 
df_2017hsML['year'] = 2017

```

```python
#### Reference https://pandas.pydata.org/pandas-docs/stable/merging.html
frames = [df_2014hsML, df_2015hsML, df_2016hsML, df_2017hsML]
df_hsML = pd.concat(frames)
```
