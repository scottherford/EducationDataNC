## 1. Merging to get the final data set. 

** Adding year column to each data set from 2014-2017.**

This is to add the year column indicating year for the machine learning ready set from 2014-2017

```python
#import required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#**********************************************************************************
# Set the directory to get ML data.You may change the directory to your own one.
#**********************************************************************************
```

```python
#**********************************************************************************
# getting the year column to each data set.
#**********************************************************************************
#2014
df_2014ML = pd.read_csv('PublicHighSchools2014_ML.csv') # read in the csv file
df_2014ML['year'] = 2014
df_2014ML.head()
df_2014ML.info(verbose=True) #to check the data type of each column
```

```python
# reading the 2015 data sets and adding year column
df_2015ML = pd.read_csv('PublicHighSchools2015_ML.csv') # read in the csv file
df_2015ML['year'] = 2015
df_2015ML.info(verbose=True) #to check the data type of each column
```

```python
# reading the 2015 data sets and adding year column
df_2016ML = pd.read_csv('PublicHighSchools2016_ML.csv') # read in the csv file
df_2016ML['year'] = 2016
df_2016ML.info(verbose=True) #to check the data type of each column
```

```python
# reading the 2015 data sets and adding year column
df_2017ML = pd.read_csv('PublicHighSchools2017_ML.csv') 
df_2017ML['year'] = 2017
df_2017ML.info(verbose=True) #to check the data type of each column
```

** Merging and creating the final data set to use.**

Since data sets are not having the same number of columns, I decide to concatenate in terms of 'outer' so that I will not lose data but instead I will create more NaN.

```python
#to merge in terms of columns, we can try con concatenating
#https://pandas.pydata.org/pandas-docs/stable/merging.html
df_final = pd.concat([df_2014ML,df_2015ML,df_2016ML,df_2017ML], axis=0, join='inner')
```

## Data Visualization of certain attributes

** 1. Compare 4 years Graduation Rate between females and males from 2014 to 2017. **


```python
df_4yrGradRate_subset =df_final[['year','GraduationRate_4yr_Female','GraduationRate_4yr_Male']]
pd.DataFrame(df_4yrGradRate_subset)

df_4yrGradRate_subset2014 = df_4yrGradRate_subset.loc[df_4yrGradRate_subset['year'] == 2014]
df_4yrGradRate_subset2014 = df_4yrGradRate_subset2014.drop(['year'], axis=1)


df_4yrGradRate_subset2015 = df_4yrGradRate_subset.loc[df_4yrGradRate_subset['year'] == 2015]
df_4yrGradRate_subset2015 = df_4yrGradRate_subset2015.drop(['year'], axis=1)


df_4yrGradRate_subset2016 = df_4yrGradRate_subset.loc[df_4yrGradRate_subset['year'] == 2016]
df_4yrGradRate_subset2016 = df_4yrGradRate_subset2016.drop(['year'], axis=1)


df_4yrGradRate_subset2017 = df_4yrGradRate_subset.loc[df_4yrGradRate_subset['year'] == 2017]
df_4yrGradRate_subset2017 = df_4yrGradRate_subset2017.drop(['year'], axis=1)
```

```python
#4 histogram plots as: x= female ,y = male
x2014 = df_4yrGradRate_subset2014['GraduationRate_4yr_Female']
y2014 = df_4yrGradRate_subset2014['GraduationRate_4yr_Male']

x2015 = df_4yrGradRate_subset2015['GraduationRate_4yr_Female']
y2015 = df_4yrGradRate_subset2015['GraduationRate_4yr_Male']

x2016 = df_4yrGradRate_subset2016['GraduationRate_4yr_Female']
y2016 = df_4yrGradRate_subset2016['GraduationRate_4yr_Male']

x2017 = df_4yrGradRate_subset2017['GraduationRate_4yr_Female']
y2017 = df_4yrGradRate_subset2017['GraduationRate_4yr_Male']

n_bins = 20
sns.set_style('whitegrid')

fig, axes = plt.subplots(2, 2,figsize=(18,16))
axes[0, 0].hist([x2014, y2014], n_bins, label=['Female', 'Male'],alpha=0.5)
axes[0, 0].set_title('4 Year GraduationRate 2014')
axes[0, 0].legend(loc=0)

axes[0, 1].hist([x2015, y2015], n_bins, label=['Female', 'Male'],alpha=0.5)
axes[0, 1].set_title('4 Year GraduationRate 2015')
axes[0, 1].legend(loc=0)

axes[1, 0].hist([x2016, y2016], n_bins, label=['Female', 'Male'],alpha=0.5)
axes[1, 0].set_title('4 Year GraduationRate 2016')
axes[1, 0].legend(loc=0)

axes[1, 1].hist([x2017, y2017], n_bins, label=['Female', 'Male'],alpha=0.5)
axes[1, 1].set_title('4 Year GraduationRate 2017')
axes[1, 1].legend(loc=0)

#sns.distplot(x,label='Female')
#sns.distplot(y,label='Male');
#plt.legend(loc=0)
#plt.show()
#sns.pairplot(df_4yrGradRate_subset,vars = ['GraduationRate_4yr_Female', 'GraduationRate_4yr_Male'],diag_kind = 'kde',palette="GnBu_d",size=5)
```
